```python
%matplotlib inline
import matplotlib
import pickle
import matplotlib.pyplot as plt

data = pickle.load(open("save.p","rb"))
rList = data["Reward_History"]

success = [sum(rList[:i]) for i in range(len(rList))]
successtofail = [(sum(rList[:i])/(i+1)) for i in range(len(rList))]

avg100 = [(sum(rList[i:i+100])/(100)) for i in range(len(rList)-100)]
```
